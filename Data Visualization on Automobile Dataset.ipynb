```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline      # this indicates print the output every time in jupyter notebook after plotting is done. No need to type plt.show()
```

```python
auto = pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

```python
# Univariate Distribution  : picking a single variable and analyzing

sns.histplot(auto['normalized_losses'],color="blue", label="100% Equities", kde=True, stat="density", linewidth=0)
plt.show()
```

```python
sns.distplot(auto['city_mpg'],color="blue", kde=True,)
plt.show()
```

```python
sns.distplot(auto['normalized_losses'], kde = False,  rug = True)
plt.show()
```

```python
# Bivariate Distribution : by using two variables

# Joint plot

sns.jointplot(auto['engine_size'], auto['horsepower'])
plt.show()
```

```python
# kind ; { 'scatter' | 'reg' | 'resid' | 'kde' | 'hex'}, optional
sns.jointplot(auto['engine_size'], auto['horsepower'], kind="kde")
plt.show()
```

```python
# Multivariate Distribution Analysis

sns.pairplot(auto[['normalized_losses',"engine_size","horsepower"]])
```

```python
# 2 Dimensional plot , one with caterogical and one with continuous

sns.stripplot(auto["fuel_type"], auto["horsepower"], jitter = True)    # ordered scatter plot
```

```python
sns.swarmplot(auto['fuel_type'], auto['horsepower'])   # no overlapping of points, better visualization for frequency
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['fuel_type'])
```

```python
# barplot : volume of things 

sns.barplot(auto['body_style'], auto['horsepower'], hue = auto["engine_location"])
```

```python
sns.countplot(auto['body_style'], hue=auto['engine_location'])
```

```python
sns.pointplot(auto['fuel_system'], auto['horsepower'], hue =auto['number_of_doors'])
```

```python
# multipanel categorical plot

sns.factorplot(x="fuel_type",
              y = "horsepower",
              hue = "number_of_doors",
              col = "engine_location",
              data = auto,
              kind = "strip")

# various types of kind input : {point, bar, count, box, violin, strip}
```

```python
# draw linear regression plot

sns.lmplot(x ="horsepower", y ="peak_rpm", data= auto, hue="fuel_type")
```
